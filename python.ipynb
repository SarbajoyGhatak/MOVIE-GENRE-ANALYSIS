```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data/movies.csv')
```

```python
df.columns
```

```python
df=df[['title','genres','budget','revenue','imdb_rating','runtime']]
```

```python
df
```

```python
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')
df['runtime'] = pd.to_numeric(df['runtime'], errors='coerce')
```

```python
df=df.dropna()
```

```python
df
```

```python
df=df[(df['budget']>0) &(df['revenue']>0)& (df['runtime']>0)]
```

```python
df
```

```python
df.isna().sum()
```

```python
df = df.reset_index(drop=True)
df.index = df.index + 1
```

```python


df.to_csv("movies_cleaned.csv", index=True)
print("✅ Cleaned dataset saved. Total rows:", len(df))
```

```python
df
```

```python

df['main_genre'] = df['genres'].apply(lambda x: x.split(',')[0].strip() if isinstance(x, str) else None)
```

```python
df
```

```python
df=df.drop(columns='genres')
```

```python
df
```

```python
df["roi"] = (((df['revenue'] - df['budget']) / df['budget']) * 100).round(2)


pd.set_option('display.float_format', '{:.4f}'.format)
```

```python
df
```

```python
df.to_csv("movies_cleaned.csv", index=False)
```

```python
df.groupby('main_genre')["roi"].mean().sort_values(ascending=False)
```
